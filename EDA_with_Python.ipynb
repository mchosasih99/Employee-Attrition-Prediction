##Univariate Analysis:

```python
df.head()
```

```python
#Separate column between numeric and categorical data type

numeric = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate',
           'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
           'PercentSalaryHike','TotalWorkingYears', 'TrainingTimesLastYear', 
           'YearsAtCompany', 'YearsInCurrentRole','YearsSinceLastPromotion', 'YearsWithCurrManager']
           
categorical = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
               'JobRole', 'MaritalStatus', 'OverTime', 'Education', 'EnvironmentSatisfaction',
               'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'PerformanceRating',
               'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance']
```

###Categorical:

```python
#Look at categorical data
df[categorical].head()
```

```python
df[categorical].shape
```

```python
#Visualize with countplot
plt.figure(figsize=(18,18))
for i in range(0, len(categorical)):
    plt.subplot(8, 3, i+1)
    sns.countplot(y=df[categorical[i]], color='green', orient='h')
    plt.tight_layout()
```

Findings:

- The target feature is imbalance, so it is necessary to over/undersampling before the ML model is carried out

###Numeric

```python
#Look at numerical data
df[numeric].head()
```

```python
df[numeric].shape
```

```python
#Visualize with boxplot
for i in range(0, len(numeric)):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y=df[numeric[i]], color='green', orient='v')
    plt.tight_layout()
```

Findings:

- Outliers are seen mainly in the MonthlyIncome, YearAtCompany, TotalWorkingYear, YearsSinceLastPromotion columns.

```python
#Visualize with violin plot

for i in range(0, len(numeric)):
    plt.subplot(4, 4, i+1)
    sns.violinplot(y=df[numeric[i]], color='green', orient='v')
    plt.tight_layout()
```

Finding:

The highly skewed columns are: MonthlyIncome, NumCompaniesWorked, PercentSalaryHike, TotalWorkingYears, YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion, YearsWithCurrManager. This column needs to be transformed first.

```python
#Visualize with distribution plot

for i in range(0, len(numeric)):
    plt.subplot(4, 4, i+1)
    sns.kdeplot(x=df[numeric[i]], color='green')
    plt.xlabel(numeric[i])
    plt.tight_layout()
```

Findings:

- YearsinCurrentRole and YearsWithCurrManager have a similar distribution of data.
- DailyRate, HourlyRate, MonthlyRate have similar data distribution too.

##Multivariate Analysis:

```python
#Plot categorical vs Target variable
categorical_clean = {"Attrition": {'No':0, "Yes": 1}}
df[categorical]=df[categorical].replace(categorical_clean)

#Visualize 1
features1 = categorical[:7]
for i in range(0, len(features1)):
    plt.subplot(2, 4, i+1) 
    sns.kdeplot(data=df, x='Attrition', hue=features1[i])
    #plt.xlabel(features1[i])
    plt.tight_layout()
```

```python
#Visualize 2
features2 = categorical[7:]
for i in range(0, len(features2)):
    plt.subplot(2, 5, i+1) 
    sns.kdeplot(data=df, x='Attrition', hue=features2[i], palette='Set1')
    plt.tight_layout()
```

Findings:

- RnD Department, Single, Travel Rarely, No Overtime, Job Involvement High, Work Life Balance Better, No Stock Option, Performance Rating Excellent, Job Level 1 (Entry Level) has a higher attrition rate than other attributes in the same feature.
- What is unique is that No Overtime and Job Involvement High actually have a big attrition.

```python
#Plot numerical data vs Attrition

numeric_discrete=['NumCompaniesWorked','TrainingTimesLastYear']
numeric_continous = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate',
                     'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany','YearsInCurrentRole',
                     'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
#Numeric 1

plt.figure(figsize=(8,8))
for i in range(0, len(numeric_discrete)):
    plt.subplot(1, 2, i+1) 
    sns.countplot(data=df, x=numeric_discrete[i], hue='Attrition')
    plt.xlabel(numeric_discrete[i])
    plt.tight_layout()
```

```python
#Numeric 2

for i in range(0, len(numeric_continous)):
    plt.subplot(3, 4, i+1) 
    sns.histplot(data=df, x=numeric_continous[i], hue='Attrition',)
    # plt.xlabel(numeric_continous[i])
    plt.tight_layout()
```

Findings:

- The higher the Job Level, Distance From Home, Work Life Balance, Age, Monthly Income, Total Working Year, Year At Company, the lower the attrition.
- At the minimum age and Monthly Income level, the attrition level is quite high.

##Correlation

```python
#Change target variable data type

df1 = df.copy()
df1['Attrition'] = df1['Attrition'].replace({'Yes':1, 'No':0})
```

```python
# Correlation heatmap
plt.figure(figsize=(15,15))
sns.heatmap(df1.corr(),
            cmap='Blues', annot=True, fmt='.2f')
```

Findings:

- No feature has a direct high correlation to Attrition.
- Distance From Home has no correlation with Age.
- Many redundant features seen from the correlation which is > .7, especially in the column containing the word 'Years'. Maybe later some features can be dropped in the next stage.
